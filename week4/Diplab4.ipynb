**TASK 1**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '/content/R.jpeg'
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
height, width, channels = image.shape
print(f"Image Dimensions: Height = {height}, Width = {width}, Channels = {channels}")
```

```python
bit_depth = image.dtype
print(f"Bit Depth: {bit_depth}")
```

```python
out_of_range_image = np.array([[300.0, -50.0], [100.0, 255.0]], dtype=np.float64)
plt.imshow(out_of_range_image)
plt.title("Out of Range Image")
plt.show()
converted_image = np.uint8(out_of_range_image)
plt.imshow(converted_image)
plt.title("Converted Image")
plt.show()
print(f"Original Image (Out of range values): \n{out_of_range_image}")
print(f"Converted Image (uint8): \n{converted_image}")
```

**TASK 2**

```python
threshold_value = 127
ret, binarized_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
plt.imshow(binarized_image)
plt.title("Binarized Image")
plt.axis('off')
plt.show()
```

**TASK 3**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
threshold_value = 127
ret, binarized_gray_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)
plt.imshow(binarized_gray_image, cmap='gray')
plt.title("Binarized Grayscale Image")
plt.axis('off')
plt.show()

```

**TASK 4**

```python
# Manual binarization using nested loops
height, width, channels = image.shape
binarized_image = np.zeros_like(image)
threshold_value = 127

for i in range(height):
  for j in range(width):

    for k in range(channels):
        if image[i, j, k] > threshold_value:
            binarized_image[i, j, k] = 255
        else:
            binarized_image[i, j, k] = 0


plt.imshow(binarized_image)
plt.title("Manual Binarized Image")
plt.axis('off')
plt.show()
```

```python








def perform_gray_scaling():

  cap = cv2.VideoCapture(0)  # 0 for default camera

  if not cap.isOpened():
    print("Cannot open camera")
    return

  while True:
    ret, frame = cap.read()

    if not ret:
      print("Can't receive frame (stream end?). Exiting ...")
      break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('Grayscale Webcam Feed', gray)

    if cv2.waitKey(1) == ord('q'):
      break

  cap.release()
  cv2.destroyAllWindows()

perform_gray_scaling()
```
